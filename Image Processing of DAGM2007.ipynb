```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
from ntpath import basename
def c(img,convert_to_rgb=False):
    if convert_to_rgb == True: img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    plt.imshow(img,cmap='gray')
def get_6_random_images(good_image_location,bad_image_location):
    bad_images = glob(bad_image_location+'/*png')
    good_images = glob(good_image_location+'/*png')
    random3bad = list(np.random.choice(bad_images,3))
    random3good = list(np.random.choice(good_images,3))
    random6images = random3good+random3bad
    return random6images
```

# Class 1

#### Let's have a look at the demo images
    - First 3 images are without spot
    - Next 3 images are with spot
    - Spots are not that noticable now

```python
random6images = get_6_random_images(bad_image_location='Class1_def/Class1_def/',good_image_location='Class1/Class1/')
```

```python
plt.figure(figsize=(10,15))
for i in range(6):
    plt.subplot(3,3,i+1)
    img=cv2.imread(random6images[i])
    c(img,convert_to_rgb=True)
plt.show()
```

#### After image processing
    - Image processing on the same images
    - After image processing the spot is very noticeable

```python
def image_processing_function_for_class_1(img):
    edges = cv2.Canny(img,50,100)
    img = cv2.merge((edges,edges,edges))
    return img
```

```python
plt.figure(figsize=(10,15))
for i in range(6):
    plt.subplot(3,3,i+1)
    img=cv2.imread(random6images[i])
    img=image_processing_function_for_class_1(img)
    c(img,convert_to_rgb=True)
plt.show()
```

#### After image processing
    - Image processing on the same images
    - After image processing: result is not as distinguishable as Class 1 but not that bad also.

```python
def image_processing_function_for_class_2(img):
    contrasted_img = cv2.addWeighted(img,2,np.zeros(img.shape,img.dtype),0,65) #65
    ret,thresh1 = cv2.threshold(contrasted_img,90,255,cv2.THRESH_BINARY) #70
    return thresh1
```

```python
plt.figure(figsize=(10,15))
for i in range(6):
    plt.subplot(3,3,i+1)
    img=cv2.imread(random6images[i])
    img=image_processing_function_for_class_2(img)
    c(img,convert_to_rgb=True)
plt.show()
```

#### After image processing
    - Image processing on the same images
    - After image processing: more noticeable

```python
def image_processing_function_for_class_3(img):
    kernel = np.ones((3,3),np.uint8)
    gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel); #c(gradient)
    ret,thresh1 = cv2.threshold(gradient,70,255,cv2.THRESH_BINARY);#c(thresh1)
    return thresh1
```

```python
plt.figure(figsize=(10,15))
for i in range(6):
    plt.subplot(3,3,i+1)
    img=cv2.imread(random6images[i])
    img=image_processing_function_for_class_3(img)
    c(img,convert_to_rgb=True)
plt.show()
```
